```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pythonda path killanimi
# \ ifadesi kod calisdiginda  acilfigi klasoru gosderir
# ornegin eger kodunuz C:\users\nicat dizisinin altindaysa szi her defasninda C:\users\nicat olarak aratirsiniz ya da C:\users\nicat 
# yada bu sekilde path yolu verisiniz 
#ancak progromalam dilerinde ozelikle pythonda ./ ifadesi ile 
        my_img = cv2.imread("./my_image.jpg"")
```

```python
print(my_img.shape)
print(f"goruntun yuksekligi(satir sayisi) = (my_img.shape[0]) ")
print(f"goruntun genasiligi(satir sayisi) = (my_img.shape[1]) ")
```

```python
my_gray_img = cv2.cvtColor(my_img,cv2.COLOR_BGR2GRAY)
print(my_gray_img.shape)
```

```python
#treshholding  siyah beyaz yani gri renkli tek kanli goruntulerini
#kenar tespiti nesen tanima nesen ayirt etme tibbi alanda hucre goruntulerini maskaleme(grount truth mask images)
#kullanilan klassik bir goruntu isleme teknigidir
#bir esik degeri sevilir 
```

```python
_, my_binary_thresh_img = cv2.threshold(my_gray_img, thresh=125, maxval=255 , type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure (figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow (my_gray_img, cmap="gray")
plt.axis("off")
plt.title=("original gray image")

plt.subplot(1,2,2)
plt.imshow (my_binary_thresh_img, cmap="gray")
plt.axis("off")
plt.title=("threshimage")
```

```python
_, my_binary_inv_thresh_img = cv2.threshold(my_gray_img, thresh=125, maxval=255 , type=cv2.THRESH_BINARY_INV)
```

```python
fig = plt.figure (figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow (my_gray_img, cmap="gray")
plt.axis("off")
plt.title=("original gray image")

plt.subplot(1,2,2)
plt.imshow (my_binary_inv_thresh_img, cmap="gray")
plt.axis("off")
plt.title=("invthreshimage")
```
